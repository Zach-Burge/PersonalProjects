```python
import numpy as np
import itertools
import pandas as pd

sudoku_df = pd.DataFrame(pd.read_csv('sudoku.csv', nrows=10**2))

def reshape(sudoku_df):
    for n in range(sudoku_df.shape[0]):
        sudoku_df.iloc[n,0] = np.reshape(list(sudoku_df.puzzle.values[n]),(9,9)).astype(int)
        sudoku_df.iloc[n,1] = np.reshape(list(sudoku_df.solution.values[n]),(9,9)).astype(int)
    return sudoku_df

def check_solved(puzzle):
    row_solved = all([all([x in puzzle[nrow,:] for x in range(1,10)]) for nrow in range(9)])
    col_solved = all([all([x in puzzle[:,ncol] for x in range(1,10)]) for ncol in range(9)])

    box_1_solved = all([x in puzzle[0:3,0:3] for x in range(1,10)])
    box_2_solved = all([x in puzzle[0:3,3:6] for x in range(1,10)])
    box_3_solved = all([x in puzzle[0:3,6:9] for x in range(1,10)])

    box_4_solved = all([x in puzzle[3:6,0:3] for x in range(1,10)])
    box_5_solved = all([x in puzzle[3:6,3:6] for x in range(1,10)])
    box_6_solved = all([x in puzzle[3:6,6:9] for x in range(1,10)])

    box_7_solved = all([x in puzzle[6:9,0:3] for x in range(1,10)])
    box_8_solved = all([x in puzzle[6:9,3:6] for x in range(1,10)])
    box_9_solved = all([x in puzzle[6:9,6:9] for x in range(1,10)])

    puzzle_solved = all([row_solved, col_solved, box_1_solved, box_2_solved, box_3_solved, box_4_solved, box_5_solved, box_6_solved, box_7_solved, box_8_solved, box_9_solved])

    if puzzle_solved:
        for line in puzzle:
            print(*line)
    return puzzle_solved

def findPossibleVals(puzzle):
    puzzle_vals = []
    for row in range(9):
        for col in range(9):
            if puzzle[row,col] == 0:
                possible_cell_vals = np.array(range(1,10))
                possible_cell_vals = np.setdiff1d(possible_cell_vals, puzzle[row,:][np.where(puzzle[row,:] != 0)]).tolist()
                possible_cell_vals = np.setdiff1d(possible_cell_vals, puzzle[:,col][np.where(puzzle[:,col] != 0)]).tolist()
            else:
                possible_cell_vals = [puzzle[row,col]]
            puzzle_vals.append(possible_cell_vals)
    return puzzle_vals

def checkSubgrid(row,col,puzzle,n):
    if row < 3:
        if col < 3:
            subgrid = n in puzzle[0:3,0:3]
        elif col < 6:
            subgrid = n in puzzle[0:3,3:6]
        else:
            subgrid = n in puzzle[0:3,6:9]
    elif row < 6:
        if col < 3:
            subgrid = n in puzzle[3:6,0:3]
        elif col < 6:
            subgrid = n in puzzle[3:6,3:6]
        else:
            subgrid = n in puzzle[3:6,6:9]
    else:
        if col < 3:
            subgrid = n in puzzle[6:9,0:3]
        elif col < 6:
            subgrid = n in puzzle[6:9,3:6]
        else:
            subgrid = n in puzzle[6:9,6:9]
    return subgrid

def solve(puzzle, puzzle_vals):
    count = 0
    solved = False
    rows = np.array(np.where(puzzle == 0))[0]
    cols = np.array(np.where(puzzle == 0))[1]
    dic = dict(zip(list(range(len(rows))), np.zeros(len(rows), dtype=int).tolist()))
    while solved == False:
        if count >= len(rows):
            solved = check_solved(puzzle)
            break
        row = rows[count]
        col = cols[count]
        len_num = len(np.array(puzzle_vals, dtype=object).reshape(9,9)[row, col])
        num = dic[count]
        while num < len_num:
            cell = np.array(puzzle_vals, dtype=object).reshape(9,9)[row,col][num]
            check_row = cell in puzzle[row,:]
            if check_row:
                num += 1
                continue
            check_col = cell in puzzle[:,col]
            if check_col:
                num += 1
                continue
            check_grid = checkSubgrid(row,col,puzzle,cell)
            if check_grid:
                num += 1
                continue
            dic[count] = num
            count += 1
            puzzle[row,col] = cell
            break
        else:
            puzzle[row,col] = 0
            dic[count] = 0
            count -= 1



# sudoku_df = reshape(sudoku_df)
# solved = check_solved(sudoku_df.solution.values[0])
# print(solved)
new_puzzle = [5,0,0,2,0,0,0,4,0,0,0,0,6,0,3,0,0,0,0,3,0,0,0,9,0,0,7,0,0,3,0,0,7,0,0,0,0,0,7,0,0,8,0,0,0,6,0,0,0,0,0,0,2,0,0,8,0,0,0,0,0,0,3,0,0,0,4,0,0,6,0,0,0,0,0,1,0,0,5,0,0]
sol = [0 for i in range(81)]
print(sol)
sudoku_df.loc[len(sudoku_df.index)] = [new_puzzle, sol]
sudoku_df = reshape(sudoku_df)
puzzle_vals = findPossibleVals(sudoku_df.puzzle.values[-1])
# print(puzzle_vals)
print(new_puzzle)
solve(sudoku_df.puzzle.values[-1], puzzle_vals)
```
